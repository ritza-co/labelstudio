```python
import pandas as pd
import spacy
from spacy.matcher import Matcher

nlp = spacy.load("en_core_web_lg")
matcher = Matcher(nlp.vocab)

pattern = [{"LOWER": "easter"}]
matcher.add("easter", [pattern])

HIGHLIGHT ='\033[91m'
END = '\033[0m'

models = ("en_core_web_sm", "en_core_web_md", "en_core_web_lg")
```

```python
df = pd.read_csv("lines_clean.csv").dropna()
texts = df['line_text']
texts = texts[texts.str.contains("Easter")]
print(texts.shape)    
```

```python

nlp_sm = spacy.load("en_core_web_sm")
nlp_md = spacy.load("en_core_web_md")
nlp_lg = spacy.load("en_core_web_lg")


docs = nlp_lg.pipe(texts)

for doc in docs:
    for sentence in doc.sents:
        matches = matcher(sentence)
        if matches:
            sent_md = nlp_md(str(sentence))
            sent_sm = nlp_sm(str(sentence))
            print_sentence_with_ent(matches, sent_sm)
            print_sentence_with_ent(matches, sent_md)
            print_sentence_with_ent(matches, sentence)
            print("---")

```
